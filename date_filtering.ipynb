**Problem 2.** [READ DATAFRAME]
- Read the combined dataset from last week into a dataframe, `df`.
- Display the top 2 rows.

```python
filename = 'scrapes-datascience.csv'
df = pd.read_csv(filename)
df.head(2)
```

**Problem 3.** [DATA CLEANING: STRING COLUMNS]
- Convert empty fields in the `parent_id` column to 'topic'.
- Replace the original `parent_id` column with the above.
- Display the top three rows.

```python
df_parent_id = df['parent_id']
df_replaced = df_parent_id.fillna('topic')
df['parent_id']=df_replaced
df.head(3)
```

**Problem 4.** [DATA CLEANING: DATES]
- Convert all dates in the `date` column to datetime's.
- Replace the original `date` column with the converted dates.
- Display the top 3 rows.

```python
df_correct_data = pd.to_datetime(df['date'])
df_correct_data
df['date'] = df_correct_data
df.head(3)
```

**Problem 5.** [DATA WRANGLING: SINGLE DATE]

- Extract all rows that fall on a Wednesday date into a dataframe `df_wed`
- Display `df_wed`

```python
wednesday = '2024-10-16'
df_wed= df[df['date'].dt.date ==pd.to_datetime(wednesday).date()]
df_wed                                   
```

**Problem 6.** [DATA WRANGLING: DATE RANGE]
- Pull all rows from Tuesday to Thursday into a dataframe `df_tuth`.
- Display `df_tuth`.

```python
tuesday_date = pd.to_datetime('2024-10-15').date()
thursday_date = pd.to_datetime('2024-10-17').date()

condition  = (df['date'].dt.date >= tuesday_date) & (df['date'].dt.date <= thursday_date)

df_tuth = df[condition]
df_tuth
```

**Problem 7.** [DATA WRANGLING: COLUMNS]
- Create a new dataframe `df_some`, based on `df_tuth` that only has the columns: `date`, `author`, `title`, and `text`.
- Display the top 3 rows of `df_some`.

```python
df_some = df_tuth[['date','author','title','text']]
df_some.head(3)
```
