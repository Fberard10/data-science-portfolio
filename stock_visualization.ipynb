**1. IMPORTING**
- Import dataframe and plot packages from lecture.

```python
import pandas as pd 
import plotly.graph_objects as go
```

**2. READING DATA FRAMES**
- Read in your stock data as data frames.
- Print the columns for each data frame.
- Print 3 rows for each data frame.

```python
df_tsla = pd.read_csv('Tesla.csv')
df_crwd = pd.read_csv('Crowdstrike.csv')

print (df_tsla.columns)
print (df_crwd.columns)

print(df_tsla.head(3))
print(df_crwd.head(3))
```

**3. DATA CLEANING**
- Convert string dates to datetimes.
- Print 3 rows for each dataframe.

```python
tsla_clean_dates = pd.to_datetime(df_tsla['date'])
df_tsla['date']  = tsla_clean_dates 

crwd_clean_dates = pd.to_datetime(df_crwd['date'])
df_crwd['date']  = crwd_clean_dates 

print(df_tsla.head(3))
print(df_crwd.head(3))
```

**4. DATA SORTING**
- Sort the dataframe by ascending dates.
- Print the top 3 rows for each dataframe.

```python
df_tsla = df_tsla.sort_values (by = 'date' , ascending = True)
df_crwd = df_crwd.sort_values (by = 'date' , ascending = True)

print (df_tsla.head(3))
print (df_crwd.head(3))
```

**5. DATA WRANGLING**
- Pull out the summer months (June 1-August 31) into new data frames.
- Print out the top 3 rows for each data frame.

```python
summer_start = pd.to_datetime('2024-06-01')
summer_end   = pd.to_datetime('2024-08-01')

tsla_summer_condition = (df_tsla ["date"] >= summer_start) & (df_tsla['date']<= summer_end)
crwd_summer_condition = (df_crwd ['date'] >= summer_start) & (df_crwd['date']<= summer_end)

df_tsla_summer = df_tsla [tsla_summer_condition]
df_crwd_summer = df_crwd [crwd_summer_condition]

print (df_tsla_summer.head(3))
print (df_crwd_summer.head(3))
```

**6. PLOT AMD**
- Plot a line graph of AMD's closing prices over the summer.

```python
fig = go.Figure()
fig.add_trace(
     go.Scatter(x = df_tsla_summer['date'], y = df_tsla_summer['close']) 
)  
fig.show()
```

**7. PLOT NVDA**
- Plot a line graph of NVDA's closing prices over the summer.

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(x = df_crwd_summer['date'], y = df_crwd_summer['close '])
)
       
fig.show()
```

**8. COMBINE PLOTS**
- Combine AMD & NVDA Plots

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(x = df_tsla_summer['date'], y = df_tsla_summer['close'])
)
fig.add_trace(
    go.Scatter(x= df_crwd_summer['date'], y = df_crwd_summer['close '],
               line = {'color' : 'purple'})
)
fig.show()
```

**9. PLOT ANNOTATIONS**
- Add Title.
- Add Axes labels.
- Rotate dates.

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(x = df_tsla_summer['date'], y = df_tsla_summer['close'], name = 'TSLA',)
)
fig.add_trace(
    go.Scatter(x= df_crwd_summer['date'], y = df_crwd_summer['close '], name = 'CRWD',
               line = {'color' : 'purple'})
)
#fig.update_layout(title = 'TSLA vs CRWD',
                  #xaxis_title = 'Date',
                  #yaxis_title = 'Closing Price')
fig.update_layout(
                title = {'text' : 'TSLA vs CRWD', 'x' :.5},
                  xaxis_title = 'Date',
                  yaxis_title = 'Closing Price',
xaxis = {'tickvals' : df_tsla_summer['date'], 'tickangle' : -75})

fig.show()
```
