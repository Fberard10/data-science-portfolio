**Question 2.** [READ DATAFRAME] 
- Print the column names and the number of rows.

```python
filename = ('scrapes-datascience.csv')
df = pd.read_csv (filename)


column_names = df.columns
number_of_rows = len(df)
print(f'''
      Column names: {column_names})
      # of rows: {number_of_rows}
''')
```

**Question 3.** [DATA WRANGLING] 
- Extract only the topic (original posts) postings into a new dataframe, `dfop`.
- Display the top 2 rows.

```python
dfop = df[df['reply'] == 0 ]
dfop.head(2)
```

**Question 4.** [BASIC STATISTICS], for `dfop`:
- Print the number of topic posts.
- Print the minimum score.
- Print the maximum score.
- Print the average score.
- Print the score standard deviations.

```python


number_of_topic_posts = len (dfop)
min_topic_post_score = min       (dfop['score'])
max_topic_post_score = max       (dfop['score'])
avg_topic_post_score = st.mean   (dfop['score'])
std_topic_post_score = st.stdev  (dfop['score'])
print (f'''
       # of posts: {number_of_topic_posts}
       Min post scores :{min_topic_post_score}
       Max post score : {max_topic_post_score}
       Avg post score : {avg_topic_post_score}
       sdv post score : {std_topic_post_score}
       ''')

```

**Question 5.** [DATA WRANGLING], for `dfop`:
- Sort the posts by the highest `score`'s descending.
- Display a dataframe with just the `author`, the `title`, and the `score` columns, for the top 5 posts.

```python
dfop_sorted = dfop.sort_values('score', ascending = False)
dfop_filtered = dfop_sorted.head(5)[['author', 'title', 'score']]
dfop_filtered.head()
```

**Question 6.** [VISUALIZATION], for `dfop`:
- Barplot the top 25 scores.

```python
top_n = 25
dfop_topn = dfop_sorted[0:top_n]

fig = px.bar(dfop_topn, x = 'title' , y = 'score')
fig.show()
```
