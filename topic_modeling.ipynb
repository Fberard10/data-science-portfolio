```python
import pandas as pd
import nltk 
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

```

**Problem 0.3. READ DATAFRAME** 
- Print the column names and the shape.

```python
folder = 'angry-society.csv'
file = 'angry-society.csv'
filepath = f'{folder}/{file}'

df = pd.read_csv(filepath)
print(df.columns)
print(df.shape)
```

**Problem 0.4. CLEAN DATA**
- Clean the `date` column 
- Clean the `parent_id` column, converting empty cells to 'TOPIC'
- Clean the `text` column, by converting empty cells to 'BLANK' 
- Display the first 3 rows of `df`

```python
df['date']   = pd.to_datetime(df['date'])
df['parent_id'] = df['parent_id'].fillna('TOPIC')
df['text']      = df['text'].fillna('BLANK')
df.head(3)

```

**Problem 0.5. ADDITIONAL DATA WRANGLING**

- Pull out the text column into the variable `text`
- Display the first 5 rows

```python
text = df['text']
text.head(5)
```

**Problem 1. DATA TRANSFORMATION**: Document Term Matrix 
- Download stopwords from the `nltk`
- Save the english stopwords in `english_stopwords` 
- Create a document-term matrix based on `text` named `dtm`
- Extract the matrix into `mat`
- Extract the column names (feature names) into `col`
- Create a dataframe of the document-term matrix named `df_dtm`
- Display the first five rows in `df_dtm`

```python
nltk.download('stopwords')
english_stopwords = stopwords.words('english')

vectorizer = CountVectorizer(stop_words= english_stopwords)
dtm = vectorizer.fit_transform(text)
mat = dtm.toarray()
col = vectorizer.get_feature_names_out()

df_dtm = pd.DataFrame(mat , columns= col)
df_dtm.head(5)
```

**PROBLEM 2. DATA MODELING**: Latent Dirichlet Allocation (LDA)

- Instantiate and LDA objects named `lda` with 5 topics (n_components), and a random state of 42, 
- Create an LDA model, `mdl`, of your `text`
- Extract the topic names into `col`
- Put the LDA model into a data frame `df_lda`
- Display the top 5 rows of `df_lda`

```python
n_topics = 5 
lda = LatentDirichletAllocation(n_components= n_topics , random_state= 42 )
mdl = lda.fit_transform(dtm)
col = lda.get_feature_names_out()
df_lda = pd.DataFrame(mdl, columns= col )
df_lda.head(5)
```

**PROBLEM 3. TEXT VISUALIZATION**: Display the top 5 words in each topic

```python
topic_word_strength = lda.components_
vocabulary = vectorizer.get_feature_names_out()
print(topic_word_strength)
print(vocabulary)
topN = 5
for topic_row in range(n_topics): 
    word_strengths = topic_word_strength[topic_row]
    indices = np.argsort(-word_strengths)
    sorted_vocabulary = vocabulary[indices]

    print(f'Topic{topic_row}: Top {topN} Words = {sorted_vocabulary[:topN]}')
```
