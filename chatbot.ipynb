```python
# Input
import os
from   dotenv import load_dotenv

# Processing (analysis)
import torch
from   transformers import pipeline
```

```python
# Check if the user has a graphics card (GPU)
device = 'cuda' if torch.cuda.is_available() else "cpu"
print(device)

# Load the HuggingFace Access Token
load_dotenv()
hf_token = os.getenv('HUGGINGFACE_WRITE_TOKEN')
```

```python
#
# Fastest way to setup a pipeline
#
ai = pipeline('text-generation', model = 'meta-llama/Llama-3.2-1B-Instruct', token = hf_token, device = device)
```

```python
context = [
    {'role': 'system', 'content' : 'You are a student.'},
#    {'role': 'user'  , 'content' : 'tell me your name?'}
]

while True:
    user_input = input('> ')
    if user_input.lower() in ['q','quit']:
        break
    user_dict = {'role'  :  'user', 'content' : user_input }
    context.append(user_dict)
    #print(context)
    response = ai(context, max_new_tokens=256, pad_token_id = ai.tokenizer.eos_token_id)
    print(f'USER: {user_input}\n')
    print(f'AI: {response[-1]['generated_text'][-1]['content']}',flush= True)
    context = response[-1]['generated_text']
```
