**Problem 2. READ DATAFRAME** 
- Print the column names and the number of rows.

```python
df = pd.read_csv('scrapes-SupermanAndLois.csv')
print (df.columns)
print(f'number of rows {len(df)}')
```

**Problem 3. CLEAN DATA**
- Clean the `date` column by converting each cell using `pd.to_datetime()`
- Clean the `parent_id` column by converting empty cells to 'topic' using `.fillna('topic')`
- Print the first 3 rows of `df`

```python
df['date'     ] = pd.to_datetime(df['date'])
df['parent_id'] = df['parent_id'].fillna('topic')
df.head(3)
```

**Problem 4. DATA WRANGLING** 
- Extract all posts (both topic and replies) for the date range October 20 - October 26 into a new dataframe, `df_week`.
- Sort `df_week` by `date`
- Display `df_week`.

```python
start_date = pd.to_datetime('2024-10-20').date()
end_date = pd.to_datetime('2024-10-26').date()

range_condition = (df['date'].dt.date >= start_date) & (df['date'].dt.date<= end_date)

df_week = df[range_condition]
df_week = df_week.sort_values(by = 'date', ascending=True)
df_week
```

**Problem 5. DATA ANALYSIS**
- Count the number of posts per day
- Print the daily_posts

```python
date_groups =  df_week.groupby(by=df['date'].dt.date)
daily_posts = date_groups.size()
daily_posts
```

**Problem 5. VISUALIZATION**, for `daily_posts`:
- Do a barplot of daily_posts

```python
fig = go.Figure()
fig.add_trace(go.Bar( 
    x = daily_posts.index,
    y = daily_posts.values
))
fig.show()
```

**Problem 7. ANALYSIS**, given `daily_posts` 
- Print the date with the most posts
- Print the number of posts on that day

```python
max_date = daily_posts.idxmax()
max_post = max(daily_posts)
print(max_date)
print(max_post) 
```

**Problem 8. ANALYSIS**
- Display the top 5 scoring posts along with author, title, and text

```python
condition = df_week['date'].dt.date == max_date
df_spike = df_week [condition]
df_spike = df_spike.sort_values(by = 'score', ascending = False)
df_top5 = df_spike.head(5)[['score', 'author', 'title', 'text' ]]
df_top5
```
