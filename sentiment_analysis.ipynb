```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

import plotly.graph_objects as go
```

**Problem 0.3. READ DATAFRAME** 
- Print the column names and the shape.

```python
folder = 'iphone-16-product-review'
filename = 'iphone-scrape.csv'
fullpath = f'{folder}/{filename}'

df= pd.read_csv(fullpath)
print(df.columns)
#print(f'# cols: {len(df.columns)}')
#print(f'# rows: {len(df)}')

df.shape
```

**Problem 0.4. CLEAN DATA**
- Clean the `date` column 
- Clean the `parent_id` column, converting empty cells to 'TOPIC'
- Clean the `text` column, by converting empty cells to 'BLANK' 
- Print the first 3 rows of `df`

```python
df['date']= pd.to_datetime(df['date'])
df['parent_id']= df['parent_id'].fillna('TOPIC')
df['text'] = df['text'].fillna('BLANK')
df.head(3)
```

**Problem 0.5. ADDITIONAL DATA WRANGLING**

- Pull out the text column into the variable `text`
- Display the first 5 rows

```python
text = df['text']
text.head(5)
```

**Problem 1. DATA ANALYSIS**: SENTIMENT 
- Do a sentiment analysis on `text` 
- Add it to the dataframe as the column `sentiment` 
- Display the first five rows in the `sentiment` column

```python
analyzer = SentimentIntensityAnalyzer()


#test = "this is a wonderful show"
#score = analyzer.polarity_scores(test)
#score['compound']

sentiments = []
for post in text:
    sentiment_group = analyzer.polarity_scores(post)
    sentiment_score = sentiment_group['compound']
    sentiments.append(sentiment_score)
 
df['sentiments'] = sentiments
df['sentiments'].head(5)


```

**PROBLEM 2. DATA ANALYSIS**: Text Visualization

- Print the top 5 negative texts and their sentiment values
- Print the top 5 postitive text and their sentiment values

```python
df_sorted = df.sort_values(by= 'sentiments')
N = 5 
df_sorted_negative_top_N = df_sorted[:N]
df_sorted_negative_top_N

df_sorted_positive_top_N = df_sorted[-N:]
df_sorted_positive_top_N = df_sorted_positive_top_N.sort_values(by = 'sentiments', ascending= False)
df_sorted_positive_top_N

print(df_sorted_negative_top_N[['text', 'sentiments']])
print(df_sorted_positive_top_N[['text', 'sentiments']])
```

**Problem 3. VISUALIZATION**, Histogram
- Visualize the distribution of sentiment as a histogram
- Add aesthetics

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x = df['sentiments'],nbinsx=50))

fig.update_layout(
                title = {'text' : 'Sentimate data for iphone ad reviews', 'x' :.5},
                  xaxis_title = 'Compound',
                  yaxis_title = 'Amount',)
                 
fig.show()
```

**Problem 4. VISUALIZATION:** Sentiment Over Time
- Sort `df` by time as `df_sorted_time`
- Visualize `df_sorted_time` as a scatter plot, for the first 100 posts
- Add aesthetics

```python
df
df_sorted_time = df.sort_values(by = 'date')
df_sorted_time

N = 100
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_sorted_time['date'][:N], y= df_sorted_time['sentiments'][:N],
              line = {'color':'purple'}
              ))
fig.update_layout(
                title = {'text' : 'Sentimate data for iphone ad reviews', 'x' :.5},
                  xaxis_title = 'Compound',
                  yaxis_title = 'Amount',)

fig.show()
```

**Problem 5. VISUALIZATION**: Pie Chart
- Assume neutral sentiment is any value between -.05 and +.05, inclusive
- Positive sentiment is any value > .05
- Negative sentiment is any value < -.05
- Draw a pie chart
- Add aesthetics

```python
neutral_condition = (df['sentiments']>= -.5) & (df['sentiments']<= .5)
postive_condition =  df['sentiments'] > .05
negative_condition = df['sentiments'] < -.05

df_neutral = df[neutral_condition]
df_positve = df[postive_condition]
df_negative = df[negative_condition]

neutrals = len(df_neutral)
positives = len(df_positve)
negatives = len(df_negative)
print(positives,neutrals, negatives)

fig = go.Figure()
fig.add_trace(
    go.Pie(labels = ["Positives", "Neutrals", "Negatives"],
           values = [positives ,    neutrals,   negatives],
           textinfo='label+percent+value')
)
fig.update_layout(
    title = 'Sentiment for iphone 16 ad reviews'
)
fig.show()
```
