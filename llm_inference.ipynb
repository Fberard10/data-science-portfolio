```python
# Input
import os
from   dotenv import load_dotenv

# Processing (analysis)
import torch # type: ignore

from   transformers import pipeline
```

```python
# Check if the user has a graphics card (GPU)
device = 'cuda' if torch.cuda.is_available() else "cpu"
print(device)

# Load the HuggingFace Access Token
load_dotenv()
hf_token = os.getenv('HUGGINGFACE_TOKEN')
```

```python
#
# Fastest way to setup a pipeline
#
ai = pipeline('text-generation', model = 'meta-llama/Llama-3.2-1B-Instruct', token = hf_token, device = device)
```

```python
call = [
    {'role': 'system', 'content' : 'You are a student.'},
    {'role': 'user'  , 'content' : 'why are lists in code important?'}
]
response = ai(call, max_new_tokens=200)
print(response[-1]['generated_text'][-1]['content'])
```
